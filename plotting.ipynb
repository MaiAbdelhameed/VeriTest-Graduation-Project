```python
import matplotlib.pyplot as plt

# Data
experiments = [
    (1, 29, 20),
    (2, 25, 21),
    (3, 10, 10),
    (4, 40, 27),
    (5, 43, 25),
    (6, 34, 26),
    (7, 64, 52),
    (8, 58, 51),
    (9, 83, 61),
    (10, 92, 66),
    (11, 80, 70),
    (12, 85, 70),
    (13, 89, 72),
    (14, 55, 52), # was 16
    (15, 94, 92), # was 17
    (16, 90, 87)  # was 18
]

# Extracting data
experiment_numbers = [exp[0] for exp in experiments if exp[1] is not None and exp[2] is not None]
train_accuracies = [exp[1] for exp in experiments if exp[1] is not None and exp[2] is not None]
test_accuracies = [exp[2] for exp in experiments if exp[1] is not None and exp[2] is not None]

plt.figure(figsize=(12, 8))
plt.scatter(train_accuracies, test_accuracies)

# Adding labels
for i, txt in enumerate(experiment_numbers):
    plt.annotate(txt, (train_accuracies[i], test_accuracies[i]), fontsize=12, ha='right')

# Adding title and labels
plt.title('Train vs Test Accuracy for Different Experiments', fontsize=16)
plt.xlabel('Train Accuracy (%)', fontsize=14)
plt.ylabel('Test Accuracy (%)', fontsize=14)

# Show plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
experiments = [
    ("KNN", 20),
    ("SVM", 27.5),
    ("KNN with new files", 51),
    ("KNN with GNN embeddings", 65)
]

# Extracting data
labels = [exp[0] for exp in experiments]
test_accuracies = [exp[1] for exp in experiments]

# Creating the plot
plt.figure(figsize=(12, 8))
plt.plot(labels, test_accuracies, 'o', markersize=10)

# Adding title and labels
plt.title('Test Accuracies for Different Experiments', fontsize=16)
plt.xlabel('Experiment', fontsize=14)
plt.ylabel('Test Accuracy (%)', fontsize=14)

# Adding data point labels
for i, txt in enumerate(labels):
    plt.annotate(txt, (i, test_accuracies[i]), fontsize=12, ha='right', xytext=(0, 5), textcoords='offset points')

# Show plot
plt.grid(True)
plt.show()
```
