```python
import matplotlib.pyplot as plt

# Data
experiments = [
    (1, 29, 20),
    (2, 25, 21),
    (3, 10, 10),
    (4, 40, 27),
    (5, 43, 25),
    (6, 34, 26),
    (7, 64, 52),
    (8, 58, 51),
    (9, 83, 61),
    (10, 92, 66),
    (11, 80, 70),
    (12, 85, 70),
    (13, 89, 72),
    (14, None, 20),
    (15, None, 27.5),
    (16, 55, 52),
    (17, 94, 92),
    (18, 90, 87)
]

# Extracting data
experiment_numbers = [exp[0] for exp in experiments if exp[1] is not None and exp[2] is not None]
train_accuracies = [exp[1] for exp in experiments if exp[1] is not None and exp[2] is not None]
test_accuracies = [exp[2] for exp in experiments if exp[1] is not None and exp[2] is not None]

# Creating the plot
plt.figure(figsize=(10, 6))
plt.scatter(train_accuracies, test_accuracies)

# Adding labels
for i, txt in enumerate(experiment_numbers):
    plt.annotate(txt, (train_accuracies[i], test_accuracies[i]), fontsize=9, ha='right')

# Adding title and labels
plt.title('Train vs Test Accuracy for Different Experiments')
plt.xlabel('Train Accuracy (%)')
plt.ylabel('Test Accuracy (%)')

# Show plot
plt.grid(True)
plt.show()
```
